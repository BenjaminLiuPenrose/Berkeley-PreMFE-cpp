```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_style('dark')
```

```python
nLs = [20, 100, 1000]
```

### Problem 7b

```python
theta = 1
numSamples = 10000
```

```python
def computeEstimators(n, numSamples = numSamples, theta = theta):
    """
    args:
    n - int, length of each sample
    numSamples - int, number of samples
    rets:
    a dict
    """
    samples = []
    theta_mles = []
    theta_mms = []
    for i in range(numSamples):
        samples.append(
            np.random.uniform(
                low = 0, 
                high = theta, 
                size = n
            )
        )
        theta_mles.append(
            np.max(
                samples[i]
            )
        )
        theta_mms.append(
            2 * np.mean(
                samples[i]
            )
        )
    
    rtn = {
        "MLE": theta_mles,
        "MM": theta_mms
    }
    return rtn
    
    
```

```python
estimatorsDict = computeEstimators(nLs[0])
thetas_mle = estimatorsDict["MLE"]
thetas_mm = estimatorsDict["MM"]
plt.figure(figsize=(10, 10))
plt.subplot(211)
plt.title("MLE dist(up) and MM estimators(down) when n is {}".format(nLs[0]))
sns.distplot(
    thetas_mle, 
    kde = True,
    rug = False,
    hist = True
)
plt.subplot(212)
sns.distplot(
    thetas_mm, 
    kde = True,
    rug = False,
    hist = True
)
mean_mle = np.mean(thetas_mle)
bias_mle = mean_mle - theta
std_mle = np.std(thetas_mle)
mean_mm = np.mean(thetas_mm)
bias_mm = mean_mm - theta 
std_mm = np.std(thetas_mm)
print("For MLE, the mean is {}, bias is {}, std is {}".format(mean_mle, bias_mle, std_mle))
print('For MM, the mean is {}, bias is {}, std is {}'.format(mean_mm, bias_mm, std_mm))
```
